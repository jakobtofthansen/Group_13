```python
# Creating a function that searches for the restaurant name, rating (bubble) and reviews count. 
# The function takes an url as input. 
def get_info(link):
    sleep(0.5)
    print("Slept 0.5, getting ", link)
    response = requests.get(link)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "lxml")
        
        N, L = [], []
        for items in soup.find_all(class_="reviewSelector"):
            try:
                name = items.find(class_="partial_entry").get_text(strip=True)
                name.items.find(class_="taLnk.ulBlueLinks").click()
                sleep(1)
                entry = name.find(class_="partial_entry").get_text(strip=True)
                #review = items.find(class_="partial_entry").get_text(strip=True)
                #entry = review.find_element_by_class_name("partial_entry")
            except:
                entry = ''
            try:
                location = items.find(class_="userLoc").get_text(strip=True)
            except:
                location = ''
            N.append(entry)
            L.append(location)
            
        return N, L

# Loops over all the links in the list of links and appends all the data to a dataframe.   
data = []
for link in links[:3]:
    n, l = get_info(link)
    data.append(pd.DataFrame({'entry': n, 'location': l}))

```

```python
trip = pd.concat(data)
trip
```
